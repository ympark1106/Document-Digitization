```python
import os
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from collections import Counter

def tokenize_sentence(sentence):
    tokens = word_tokenize(sentence)
    return tokens

nltk.download('punkt')
nltk.download('stopwords')

def remove_stopwords(tokens):
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
    return filtered_tokens

corpus_folder_path = 'corpus_number_removed'

txt_files = []
for file_name in os.listdir(corpus_folder_path):
    if file_name.endswith('.txt'):
        txt_files.append(os.path.join(corpus_folder_path, file_name))

thesaurus_data = {}
for txt_file in txt_files:
    with open(txt_file, 'r', encoding='utf-8') as file:
        corpus_text = file.read()
    sentences = sent_tokenize(corpus_text)  

    for sentence in sentences:
        sentence_without_periods = sentence.replace('.', '')  
        sentence_lower = sentence_without_periods.lower() 
        tokens = tokenize_sentence(sentence_lower)
        filtered_tokens = remove_stopwords(tokens)
        
        window_size = len(filtered_tokens) - 1 
        for i in range(len(filtered_tokens)):
            current_word = filtered_tokens[i]
            context_words = filtered_tokens[:i] + filtered_tokens[i+1:]
            for j, context_word in enumerate(context_words):
                weight = 1 / (j + 1) 
                if current_word not in thesaurus_data:
                    thesaurus_data[current_word] = Counter({context_word: weight})
                else:
                    thesaurus_data[current_word].update({context_word: weight})

output_file_path = '__.txt'
with open(output_file_path, 'w', encoding='utf-8') as file:
    for current_word in thesaurus_data:
        synonyms = thesaurus_data[current_word]
        top_synonyms = sorted(synonyms.items(), key=lambda x: x[1], reverse=True)[:10]
        line = f"{current_word}: "
        for synonym, weight in top_synonyms:
            line += f"{synonym} ({weight}), "
        line = line.rstrip(", ") + "\n"
        file.write(line)

print("  .")
```

```python
import os
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from collections import Counter

#  
def tokenize_sentence(sentence):
    tokens = word_tokenize(sentence)
    return tokens

nltk.download('punkt')
nltk.download('stopwords')

unwanted_words = ["shall", "must", "may", "otherwise", "new", "ii", "iii", "iv", "iiia", "iia", "vi", "iiib", "vii", "viii", "ipa", "iva", "name", "contractor"]

def remove_stopwords(tokens):
    stop_words = set(stopwords.words('english'))
    stop_words.update(unwanted_words)  #     
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
    return filtered_tokens

corpus_folder_path = 'corpus_number_removed'

txt_files = []
for file_name in os.listdir(corpus_folder_path):
    if file_name.endswith('.txt'):
        txt_files.append(os.path.join(corpus_folder_path, file_name))

thesaurus_data = {}
for txt_file in txt_files:
    with open(txt_file, 'r', encoding='utf-8') as file:
        corpus_text = file.read()
    sentences = sent_tokenize(corpus_text)

    for sentence in sentences:
        sentence_without_periods = sentence.replace('.', '')  
        sentence_lower = sentence_without_periods.lower()
        tokens = tokenize_sentence(sentence_lower)
        filtered_tokens = remove_stopwords(tokens)
        
        window_size = len(filtered_tokens) - 1  
        for i in range(len(filtered_tokens)):
            current_word = filtered_tokens[i]
            context_words = filtered_tokens[:i] + filtered_tokens[i+1:]
            for j, context_word in enumerate(context_words):
                weight = 1 / (j + 1)  #      
                if current_word not in thesaurus_data:
                    thesaurus_data[current_word] = Counter({context_word: weight})
                else:
                    thesaurus_data[current_word].update({context_word: weight})

output_file_path = '___.txt'
with open(output_file_path, 'w', encoding='utf-8') as file:
    for current_word in thesaurus_data:
        synonyms = thesaurus_data[current_word]
        top_synonyms = sorted(synonyms.items(), key=lambda x: x[1], reverse=True)[:10]
        line = f"{current_word}: "
        for synonym, weight in top_synonyms:
            line += f"{synonym} ({weight}), "
        line = line.rstrip(", ") + "\n"
        file.write(line)

print("  .")
```

#  Word Network 

```python
import networkx as nx
from pyvis.network import Network

new_thesaurus_file = '_800_.txt'

G = nx.Graph()

with open(new_thesaurus_file, 'r', encoding='utf-8') as file:
    for line in file:
        line = line.strip()
        word, synonyms_str = line.split(': ', 1)
        synonyms_list = synonyms_str.split(', ')
        G.add_node(word)  #   
        for synonym in synonyms_list:
            synonym_word, similarity = synonym.split('(')
            similarity = float(similarity[:-1])
            G.add_edge(word, synonym_word, weight=similarity)  #     

layout = nx.spring_layout(G, seed=42)  #     

nt = Network(notebook=True, height='800px', width='100%')

for node in G.nodes():
    nt.add_node(node, label=node, title=node)

for edge in G.edges():
    source, target, weight = edge[0], edge[1], G.edges[edge]['weight']
    nt.add_edge(source, target, value=weight, title=f'Weight: {weight}')

output_file = "word_thesaurus_network_800.html"
nt.show(output_file)
```
